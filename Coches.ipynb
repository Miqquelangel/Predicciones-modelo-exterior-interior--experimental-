#MODELO BÁSICO DE DIFERENCIACIÓN ENTRE EXTERIOR E INTERIOR DE UN COCHE.

##EJECUTAR LA PRIMERA CELDA

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
!wget https://raw.githubusercontent.com/Miqquelangel/Portfolio-Miguel-Angel/main/Imagen/coc.h5
model = tf.keras.models.load_model("/content/coc.h5")
def Prediccion_imagen(file): 
  img = tf.io.read_file(file)
  img = tf.image.decode_image(img)
  img = tf.image.resize(img, [224, 224])
  img = img/tf.math.reduce_max(img)
  model.predict(tf.expand_dims(img, axis=0)) # Expandimos una dimensión por que nuestro modelo predice el batch_size
  pred = model.predict(tf.expand_dims(img, axis=0))
  clases= np.array(["Interior", "Exterior"])
  predclass = clases[int(tf.round(pred))]
  plt.figure(figsize = (10,10))
  plt.imshow(img)
  plt.title(f"Predicción: {predclass}")
  plt.axis(False);
```
